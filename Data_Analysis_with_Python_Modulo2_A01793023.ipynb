## Python Packages for Data Science
That dataset description itself tells little about the analysis; is just a list of probable attributes but the data nor the insights are there; we will call this in one minute but first let's talk about useful data science packages in python.

Packages are libaries or repositories containning a variety of functions, classes and methods that are reusable code for a certain task; for data analysis we have:

* Scientific computing  libraries such as:
 - Pandas: It offer datastructures and tools for effective manipulation and analysis; fast access to data and the primary tool is the DataFrame that is a 2x2 Matrix where columns are attributes and rows are observations.
 - NumPy: Helpful for using arrays and matrices (and their respective math)
 - SciPY: includes functions for some advanced math problems like intetgrals, solving differential equations, optimization
* Visualization libraries susch as:
 - Matplotlib: for creating plots and graphs, this is the most.popular visualization tool in Python
 - Seaborn: is based on Matplotlib as a highlevel tool for other charts like heatmaps, time series, violin plots.
* Algorithmic libraries, such as:
 - Scikit-learn: stastical modeling from transformations to models used in regrression, classification, clustering and so on
 - Statsmodels: useful for estimate statistical models, statisical tests and explore data.

## Getting Started Analyzing Data in Python
After we have called our pandas DataFrame we can use multiple methods to extract initial information of the data we have there.

Pandas has several data types, it is useful to be fammiliar with them, those types are:

| Pandas Type | Native Python Type |  Description |
|--- | --- | --- |
| object | string | numbers and strings|
| int64 | int | integer numbers |
| float64 | float | real numbers (floats) |
| datetime64, timedelta(ns) | datetime | time data |

Not all functions are applicable to all data types; we cannot do math with all and we cannot, for example, extract days from a non time value. This is why these types are important.

To get the datatypes of the data one can use the method:

``` python
df.dtypes
```
This returns a list of the attributes and the Pandas Type of each column.

Other way to explore the data is to get quick statistics with the method describe; by default this returns only something for the numerical values but it can do it also for object type, we do this wth:

``` python
df.describe(include='all')
```
| Staistic | Description |
|--- | --- |
| count | number of items |
| unique* | number of unique values |
| top* | most frequently recurring object (mode of object) |
| freq* | number of times the top values (top) appears in the column|
| mean | average of numeric values|
| std | standard deviation of values |
| min | min of values |
| 25% | first quartile of values|
| 50% | median of vlaues |
| 75% | third quartile of values |
| max | max of values |

\* only for objects

We can use this to have a first impression of the distribution of each attribute and some of the values for objects.

Finally another useful method is:

``` python
df.info()
```

This one returns first and last rows to look at the dataframe
