## Exploratory Data Analysis
Exploratory Data Analysis or in short EDA is a prelimary step to:
* Summarize main characteristics of the data
* Gain better understanding of the data set
* Uncover relationships between variables
* Extract important variables

The main question is: "What are the characteristics that have the most impact on a variable"

For this we use:

* Descriptive Statistics - Basic features
* Groupping data with GroupBY
* Analysis of variance (ANOVA)
* Correlation between variables


## Descriptive Statistics
Descriptive statistics helps to describe basic features of the dataset and obtains a short summary of the data. For this we can use:

``` python
import pandas as pd
df.describe()
```
This shows the mean, sd, quartile, extreme values and give us an idea of the distribution of the variables; usually used only for numeric values.

For categories we can use
``` python
import pandas as pd
df['column'].value_counts()
```
This give us the unique values and how many of those we observe.

Other way to visualize numeric data is with the boxplots where we can find:
* Median
* Q3
* Q1
* IQR range for Upper and Lower (1.5*IQR + [Q1, Q3])
* We can also visualize outliers as dots

``` python
import pandas as pd
import seaborn as sns
sns.boxplot(x='variable', y='variable2', data=df)
```

Other thing we can use is scatter plots where we see the relationship between two numeric values.
This can help to see if a variable (X) can explain another variable (Y).

``` python
import pandas as pd
import seaborn as sns
x=df['column1']
y=df['target_column']

plt.scatter(x,y)
```



## GroupBy in Python
Groupping can help to use categorical values (or numerical through bins) against other variables, usually against a numeric where we can compute statistics like sum, count, std, etc.

As an example:
We can try to find the average value of a numeric value against a category
 
``` python
import pandas as pd

df_test = df[['category1', 'numeric1']]

df_test.groupby(['category1']), as_index=False).mean()
```

We are not limited to one category with this technique, we can use more than that.

We can also incorporate the pivot method to see this better

``` python
import pandas as pd

df_test = df[['category1', 'category2 'numeric1']]

df_grp = df_test.groupby(['category1', 'category2']), as_index=False).mean()

df_pivot = df_grp.pivot(index='category1', columns='category2')
```

This makes the observations of category 2 columns and category 1 rows, the numeric values will be used across this grid to reflect what we are looking for.


Another technique se can use is heatmaps to see relationships between variables and their impact.

``` python
import pandas as pd
from matplotlib import pyplot as plt
# .... code above
plt.pcolor(df_pivot, cmap='RdBlu')
plt.colorbar()
plt.show()
```



## Analysis of Variance ANOVA
Analysis of variance or ANOVA helps to analyze categorical variables and their correlation among different categories.

ANOVA is an statistical test can be used for finding correlation between different groups of a categorical variable. ANOVA return us if there are statistical significant differences across the categorical values like between A and B.

ANOVA returns a F-Test Score that is the variation between sample group means divided by variation within sample group. P-value: provide us a confidence degree of the test.

image.png

In the image above we can see an intuitive example of the ANOVA analysis where difference is quite signifcant between the two groups. Here is an example in python for the car dataset:

``` python
import pandas as pd
import stats

df_anova = df[['make', 'price']]
grouped_anova= df_anova.groupby[['make']]
anova_results = stats.f_oneway(grouped_anova.get_group('honda')['price'], grouped_anova.get_group('jaguar')['price'])

```

This wuld return the ANOVA statistics, we usually look here for small p-values <=0.05 and large F

## Correlation
Correlation is an statistic is an statistical metric for measuring to what extent different variables are interdependent, in other words: how changes in one variable affect the other one; is in the same direction? is in opposite direction? or there is no response from the other one at all?

For example:
* Lung cancer is correlated to smoking
* pressence of umbrellas are correlated to rain

**Correlation doesn't imply causation**

We can see this with an scatterplot in python with regression lines with:

``` python
import pandas as pd
import seaborn as sns
sns.regplot(x='category1', y='numeric1', data=df)
plt.ylim(0,)
```
The code above will produce an scatter plot and a regression line that shows the relationship or in this case correlation between the two variables, which again can be:

* Positive Correlation > 0
* Negative Correlation< 0
* Null Correlation = 0

## Correlation - Statistics
One method to measure the strength of correlation between two features is through Pearson Correlation (linear), it will give you the correlation coefficient and a p-value.

**Correlation coefficient:**
* Close to +1: Large positive relationship
* Close to -1: Large negative relationship
* Close to 0: No relationship

**p-value**
* P-value < 0.001 **Strong** certaintiy in the result
* P-value < 0.05 **Moderate** certainty in the result
* P-value < 0.1 **Weak** certainty in the result
* P-value > 0.1 **No** certainty in the result

Here is a graphic representation of different correlation values

image.png

We can get these values as follows
``` python
import pandas as pd
import stats

pearson_coef, p_value = stats.pearsonr(df['col1'], df['col2'])
```
This will return both the coefficient and the p_vlaue and we can use those values to determine the relationship and confidence of the statistic.

We can actually also extend this to all variables with techniques like a correlation heatmap, where we can see all the variables croessed agains each other in a matrix-shape of the features and how they behave against each other in terms of correlation.
